```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl = 'ko', tz = 540)
```

```python
kw_list = ['가평빠지','강릉해수욕장','대천해수욕장','양양해수욕장']
```

```python
pytrends.build_payload(kw_list, cat = 0,
                       timeframe='2024-08-01 2024-08-15',
                       geo='KR', gprop='')
```

```python
data = pytrends.interest_over_time()
#pytrendss.multirange_interest_over_time()
```
